## Using a neural network to fit the data


 **What Are Neural Networks?**
- Neural networks are mathematical models used in deep learning to approximate complex functions.
- While inspired by biological neurons, modern artificial neural networks (ANNs) are only loosely related to how the human brain works.
- These networks use a combination of linear and nonlinear functions to process information.

**Understanding Artificial Neurons**
- The basic unit of a neural network is the **neuron**.
- Each neuron performs two main operations:
  1. **Linear Transformation**: It takes an input \( x \), scales it by a weight \( w \), and adds a bias \( b \).

    

  2. **Nonlinear Activation Function**: The result is passed through a function \( f \), such as **tanh** or **ReLU**, to introduce nonlinearity.

    

- The activation function allows the network to capture complex patterns.



**Building a Multilayer Neural Network**
- A single neuron can only represent simple relationships.
- To model more complex functions, neurons are stacked into **layers**.
- The output of one layer serves as the input to the next:



 

- When multiple layers are combined, we get a **deep neural network (DNN)**.

**Understanding the Error Function**
- The goal of training a neural network is to minimize the **error** between predicted and actual outputs.
- Unlike simpler models, neural networks have **non-convex** error functions, meaning there is no single "best" solution.
- The loss function (e.g., mean squared error) measures how far the predicted output is from the ground truth.

**Training the Neural Network**
1. **Forward Pass**: Inputs are processed through the layers to generate predictions.
2. **Error Calculation**: The loss function computes the difference between predictions and actual values.
3. **Backward Pass (Backpropagation)**:
   - The model adjusts weights and biases using **gradient descent** to reduce the error.
4. **Iteration**: Steps 1-3 are repeated until the model reaches a satisfactory level of accuracy.

###  Activation Functions in Neural Networks

***Neural Networks: The Basics***
- A neural network consists of **linear operations** (like scaling and offset) followed by **activation functions**.
- The linear operation applies weights and biases to inputs, but it alone cannot model complex patterns.

***Why Do We Need Activation Functions?***
Activation functions serve two main purposes:
1. **Inside the network**  They allow different slopes at different input values, enabling the network to approximate complex functions.
2. **At the final layer**  They constrain outputs to a defined range (e.g., between 0 and 1 for classification problems).

***Example: Assigning a "Good Doggo" Score**
- Suppose we want to rate images based on how "dog-like" they are.
- Some categories:
  - **High scores**  Retrievers.
  - **Medium scores**  Bears (vaguely dog-like).
  - **Low scores**  Garbage trucks, airplanes.
- The issue: Without activation functions, the model's scores can be unpredictable (e.g., scoring a dog **12/10** or a bear **-10**).



***Capping the Output Range***
To prevent extreme values, we use activation functions that **restrict the range** of outputs:
- **HardTanh**: Clips outputs within a fixed range (e.g., -1 to +1 or 0 to 10).
  - Example: If an output is **12**, it gets capped at **10**.

***Compressing the Output Range**
Other activation functions smoothly transform outputs into a **bounded range**:
- **Sigmoid**: Outputs between **0 and 1**.
- **Tanh**: Outputs between **-1 and 1**.
  - Example:
    - **Garbage truck**  -0.97 (definitely not a dog).
    - **Bear**  ~-0.3 (somewhat dog-like).
    - **Dog**  ~0.99 (clearly a dog).
- **Benefit**: The network is sensitive to meaningful input differences while ignoring extreme variations.

**Sensitivity to Inputs**
- The activation function makes **small changes matter**:
  - A **polar bear** might score higher than a **grizzly bear** due to a more canine-like face.
  - A **koala bear** might score lower because it looks less like a dog.
  - A **garbage truck** will always score low, no matter how we tweak the input.




### Different Types of Activation Functions
There are several types of activation functions, each with different properties. Some common ones include:  

- **Smooth Functions:**  
  - **Tanh**: Outputs between -1 and 1.  
  - **Softplus**: Similar to ReLU but differentiable everywhere.  

- **Hard (Piecewise-Linear) Functions:**  
  - **HardTanh**: Similar to Tanh but with sharp cutoffs.  
  - **ReLU (Rectified Linear Unit)**: Outputs 0 for negative inputs and passes positive inputs unchanged.  
  - **LeakyReLU**: A modified version of ReLU that allows a small slope for negative values to avoid zero gradients.



- **Sigmoid Function:**  
  - Used in early deep learning models.  
  - Still useful for probability-based outputs since it maps values to the **0-1** range.


#### Choosing the Best Activation Function
Neural networks rely on activation functions to introduce non-linearity and make learning possible. The best choice depends on the problem.  

- **Key Properties of Activation Functions:**  
  1. **Non-linearity**: Without it, stacking layers would still result in a linear model.  
  2. **Differentiability**: Allows gradients to be computed for backpropagation.  
  3. **Sensitive Range**: Input changes must lead to meaningful output changes for effective learning.  
  4. **Saturation Effect**: Some functions have an insensitive range (e.g., Sigmoid, Tanh), where large input changes result in little or no output change.  

- **Other Considerations:**  
  - Some functions have **bounded output ranges** (e.g., Sigmoid, Tanh) while others do not (e.g., ReLU).  
  - Functions like ReLU and LeakyReLU are often preferred for deep networks because they avoid vanishing gradients in the sensitive range.  


### How Activation Functions Help Learning*
- **Different units respond to different input ranges.**  
- **Errors backpropagate effectively only through neurons in the sensitive range.**  
- **For neurons in the saturated range (flat areas of activation function), gradients are near zero, making learning slow.**  
- **Training behaves similarly to linear regression in the sensitive range, making optimization easier.**  


### Why Stacking Many Activation Units Works 
- When **multiple layers of linear + activation units** are combined, the network gains the power to **approximate complex functions**.  
- **Different neurons activate for different inputs**, creating a flexible and learnable system.  
- **Gradient descent optimization works well in the sensitive range**, improving model learning.

### Understanding Learning in Neural Networks

**What Does Learning Mean for a Neural Network?**  
- Neural networks are **universal approximators**  they can model complex, highly non-linear processes.  
- They consist of **stacks of linear transformations followed by activation functions**, allowing them to capture intricate patterns.  
- Through **gradient descent**, we can efficiently estimate millions of parameters.  
- The advantage? **We dont need to explicitly define the function** (e.g., quadratic, polynomial). Instead, **the model learns it from data**.  

**How Neural Networks Learn to Model Complex Functions**  
- Consider **neurons with the Tanh activation function**:  
  - Each neuron has a unique **weight** and **bias**, controlling the center and shape of the curve.  
  - These small building blocks combine to form more complex behaviors.  
- **Example: Combining Neurons**  
  - **Adding two neurons (A + B, C + D)** creates new curves.  
  - **Stacking two layers of neurons** introduces even more complex patterns.  
  - Even with just **four neurons in two layers**, we see intricate behaviors like multiple peaks and bumps.  



**Training a Neural Network**  
- Training is the process of **adjusting weights and biases** to make the network solve a task (e.g., predicting temperature).  
- The goal is to make the model generalize:  
  - It should **perform well on unseen data** generated by the same process as the training data.  
  - The network **discovers the structure of the data** through its learned numerical representations.  

**Learning Without Explicit Models**  
- Traditional models (e.g., physics-based) rely on **assumed equations** to describe the world.  
- Example:  
  - If we assume **temperature measurements are linear**, we can only model straight-line relationships.  
- **Why is this a limitation?**  
  - Real-world data is rarely simple.  
  - Many relationships are **too complex to describe with a fixed equation**.  

**Deep Learning: A Data-Driven Approach**  
- Instead of **explicitly defining the model**, deep learning learns from examples:  
  - **We provide inputs and outputs.**  
  - **We define a loss function.**  
  - **The network adjusts its parameters through backpropagation.**  
- **Key advantage:** It can handle **high-dimensional, complex relationships** without requiring predefined formulas.  
- **Key tradeoff:** We may not always understand why it works, but it allows us to solve problems that were previously infeasible.

### Understanding the PyTorch `nn` Module


**Introduction to PyTorch `nn` Module**  
- **Why use PyTorchs `nn` module?**  
  - It provides pre-built **building blocks** for neural networks.  
  - It helps us scale up from simple models (like linear regression) to **deep learning architectures**.  

- **What is `torch.nn`?**  
  - It is a **submodule** in PyTorch designed for creating neural networks.  
  - It contains **modules** (also called **layers** in other frameworks).  

- **What is a PyTorch Module?**  
  - A **module** in PyTorch is a **Python class** that inherits from `nn.Module`.  
  - It contains **parameters** (e.g., weights and biases) that get optimized during training.  
  - It can also have **submodules** (other `nn.Module` instances).  

>  **Important:** Submodules **must** be defined as **top-level attributes** (not inside a `list` or `dict`). Otherwise, the optimizer wont be able to track them.  
> - If you need a **list or dictionary of submodules**, use `nn.ModuleList` or `nn.ModuleDict`.  


### Using `nn.Linear` to Replace a Linear Model
- **What is `nn.Linear`?**

### __call__

 In Python, the __call__ method allows an instance of a class to be called like a function.

```python
class Greeter:
    def __init__(self, name):
        self.name = name

    def __call__(self):
        print(f"Hello, {self.name}!")

g = Greeter("John")
g()  # Output: Hello, John!
```

### Understanding `__call__` vs. `forward` in `nn.Module`
- **Calling a PyTorch model:**  
   - `__call__` **automatically calls `forward`** while handling additional functionalities.  
  - Calling `forward` directly **skips important pre- and post-processing steps**, causing potential errors.  


```python
y = model(x)  # Correct way
y = model.forward(x)  # Silent error! Dont do this
```

**Training Loop in PyTorch**  
- **Training Code:**

```python
def training_loop(n_epochs, optimizer, model, loss_fn, t_u_train, t_u_val, t_c_train, t_c_val):
    for epoch in range(1, n_epochs + 1):
        t_p_train = model(t_u_train)
        loss_train = loss_fn(t_p_train, t_c_train)

        t_p_val = model(t_u_val)
        loss_val = loss_fn(t_p_val, t_c_val)

        optimizer.zero_grad()  # Reset gradients
        loss_train.backward()  # Compute gradients
        optimizer.step()  # Update parameters

        if epoch == 1 or epoch % 1000 == 0:
            print(f"Epoch {epoch}, Training loss {loss_train.item():.4f}, Validation loss {loss_val.item():.4f}")
```

- **Changes Compared to Earlier Manual Implementation:**
  - We **pass the entire model** instead of manually specifying weights and biases.
  - The model holds its parameters internally and provides them to the optimizer.
  - **The training loop remains mostly the same.**

**Using Built-in Loss Functions in PyTorch**  
- **PyTorch provides common loss functions** under `torch.nn`.  
- Instead of manually defining Mean Squared Error (MSE), we use `nn.MSELoss()`.



- **Final Model Training:**

```python
import torch
import torch.nn as nn
import torch.optim as optim

# -----------------------------
# 1. Sample Data
# -----------------------------
# Input temperatures in Fahrenheit
t_u = torch.tensor([35.7, 55.9, 58.2, 48.9, 33.9,
                    21.8, 48.4, 60.4, 68.4, 20.0], dtype=torch.float32)

# Target temperatures in Celsius
t_c = torch.tensor([0.5, 13.5, 14.0, 11.0, 6.8,
                   -4.3, 11.8, 15.3, 20.0, -6.7], dtype=torch.float32)

# Normalize input (helps convergence)
t_u_norm = t_u / 100.0

# Split into train/val
t_u_train = t_u_norm[:8].unsqueeze(1)  # shape (8, 1)
t_u_val = t_u_norm[8:].unsqueeze(1)    # shape (2, 1)
t_c_train = t_c[:8].unsqueeze(1)
t_c_val = t_c[8:].unsqueeze(1)

# -----------------------------
# 2. Training Loop
# -----------------------------
def training_loop(n_epochs, optimizer, model, loss_fn,
                  t_u_train, t_u_val, t_c_train, t_c_val):
    for epoch in range(1, n_epochs + 1):
        model.train()
        train_pred = model(t_u_train)
        train_loss = loss_fn(train_pred, t_c_train)

        optimizer.zero_grad()
        train_loss.backward()
        optimizer.step()

        model.eval()
        with torch.no_grad():
            val_pred = model(t_u_val)
            val_loss = loss_fn(val_pred, t_c_val)

        if epoch % 500 == 0 or epoch == 1:
            print(f"Epoch {epoch:5d}: Train loss = {train_loss.item():.4f}, Val loss = {val_loss.item():.4f}")

# -----------------------------
# 3. Model, Optimizer, Loss
# -----------------------------
linear_model = nn.Linear(1, 1)
optimizer = optim.SGD(linear_model.parameters(), lr=1e-2)
loss_fn = nn.MSELoss()

# -----------------------------
# 4. Run Training
# -----------------------------
training_loop(
    n_epochs=3000,
    optimizer=optimizer,
    model=linear_model,
    loss_fn=loss_fn,
    t_u_train=t_u_train,
    t_u_val=t_u_val,
    t_c_train=t_c_train,
    t_c_val=t_c_val
)

# -----------------------------
# 5. Show Results
# -----------------------------
print("\nFinal Model Parameters:")
print(f"Weight: {linear_model.weight.item():.4f}")
print(f"Bias: {linear_model.bias.item():.4f}")

# Example prediction
example_f = torch.tensor([[50.0]]) / 100.0
predicted_c = linear_model(example_f)
print(f"\nPredicted Celsius for 50F: {predicted_c.item():.2f}C")
```

- The **loss decreases over time**, indicating that the model is learning.  
- The **final weight and bias values** define the learned transformation.

### Building a Neural Network in PyTorch 

**Transitioning from Linear Models to Neural Networks**  
- We've learned how to train a **linear model** using PyTorch, but now we take the next step:  
  **Replacing it with a neural network**.  
- In this case, since our problem is fundamentally linear, a neural network wont necessarily improve the model.  
- However, **learning to move from linear models to neural networks** in a simple case helps us later when handling more complex problems.  

**Defining a Neural Network with `nn.Sequential`**  
- A simple neural network consists of:  
  1. A **hidden layer** (linear transformation).  
  2. An **activation function** (Tanh, ReLU, etc.).  
  3. An **output layer** (another linear transformation).  

- **PyTorch `nn.Sequential` allows stacking layers easily:**

```python
import torch.nn as nn

seq_model = nn.Sequential(
    nn.Linear(1, 13),  # Input Layer  Hidden Layer
    nn.Tanh(),         # Activation Function
    nn.Linear(13, 1)   # Hidden Layer  Output Layer
)

print(seq_model)
```

- This model:
  - Takes **1 input feature**.
  - Expands it to **13 neurons** in the hidden layer.
  - Applies **Tanh activation** (introduces non-linearity).
  - Combines the **13 features into 1 output**.

**Inspecting Model Parameters**  
- We can check the **weights and biases** of our model: 

```python
[param.shape for param in seq_model.parameters()]
```

- **Explanation:**
  - `torch.Size([13, 1])`  **Weights of first linear layer** (`1 input  13 neurons`).
  - `torch.Size([13])`  **Biases of first linear layer**.
  - `torch.Size([1, 13])`  **Weights of output layer** (`13 neurons  1 output`).
  - `torch.Size([1])`  **Bias of output layer**.

- **We can also check parameter names for better clarity:**

```python
for name, param in seq_model.named_parameters():
    print(name, param.shape)
```

**Assigning Meaningful Names to Layers**  
- Instead of numbered layers, we can use an **OrderedDict** for clarity:

```python
from collections import OrderedDict

seq_model = nn.Sequential(OrderedDict([
    ('hidden_linear', nn.Linear(1, 8)),      # First linear layer
    ('hidden_activation', nn.Tanh()),       # Activation function
    ('output_linear', nn.Linear(8, 1))      # Output layer
]))

print(seq_model)
```

```python
for name, param in seq_model.named_parameters():
    print(name, param.shape)
```

### **5. Training the Neural Network**  
- **Defining Optimizer and Training:**

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Sample Data
#t_c = [0.5, 14.0, 15.0, 28.0, 11.0, 8.0, 3.0,-4.0, 6.0, 13.0, 21.0]
#t_u = [35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4]
t_u = torch.tensor([35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4], dtype=torch.float32)
t_c = torch.tensor([0.5, 14.0, 15.0, 28.0, 11.0, 8.0, 3.0,-4.0, 6.0, 13.0, 21.0], dtype=torch.float32)

# Normalize input
t_u_norm = t_u / 100.0

# Train/val split
t_u_train = t_u_norm[:8].unsqueeze(1)
t_u_val = t_u_norm[8:].unsqueeze(1)
t_c_train = t_c[:8].unsqueeze(1)
t_c_val = t_c[8:].unsqueeze(1)

# Sequential Model with tanh activation
from collections import OrderedDict

seq_model = nn.Sequential(OrderedDict([
    ('hidden_linear', nn.Linear(1, 8)),      # First linear layer
    ('hidden_activation', nn.Tanh()),       # Activation function
    ('output_linear', nn.Linear(8, 1))      # Output layer
]))

# Optimizer
optimizer = optim.SGD(seq_model.parameters(), lr=1e-3)

# Training loop function
def training_loop(n_epochs, optimizer, model, loss_fn,
                  t_u_train, t_u_val, t_c_train, t_c_val):
    for epoch in range(1, n_epochs + 1):
        model.train()
        train_pred = model(t_u_train)
        train_loss = loss_fn(train_pred, t_c_train)

        optimizer.zero_grad()
        train_loss.backward()
        optimizer.step()

        model.eval()
        with torch.no_grad():
            val_pred = model(t_u_val)
            val_loss = loss_fn(val_pred, t_c_val)

        if epoch % 500 == 0 or epoch == 1:
            print(f"Epoch {epoch:5d}: Train loss = {train_loss.item():.4f}, Val loss = {val_loss.item():.4f}")

# Run training
training_loop(
    n_epochs=5000,
    optimizer=optimizer,
    model=seq_model,
    loss_fn=nn.MSELoss(),
    t_u_train=t_u_train,
    t_u_val=t_u_val,
    t_c_train=t_c_train,
    t_c_val=t_c_val
)
```

- **Checking Outputs and Gradients:**

```python
print('Output:', seq_model(t_un_val))
print('Answer:', t_c_val)
print('Hidden layer gradients:', seq_model.hidden_linear.weight.grad)
```

- The **loss decreases** over epochs, showing that the model is learning.
- **Hidden layer gradients** help us inspect learning behavior.

**Comparing Neural Network vs. Linear Model**  
- **Plot Predictions to See the Difference:**  

```python
import matplotlib.pyplot as plt
import torch

# Generate a range of Fahrenheit values
t_range = torch.arange(20., 90.).unsqueeze(1)  # shape (70, 1)

# Plot settings
plt.figure(figsize=(8, 5))
plt.xlabel("Fahrenheit")
plt.ylabel("Celsius")
plt.title("Fahrenheit to Celsius: Actual vs. Predicted")

# Actual data points
plt.plot(t_u.numpy(), t_c.numpy(), 'o', label="Actual Data")

# Model predictions on a smooth range (scaled input)
predicted_curve = seq_model((t_range / 100.0)).detach().numpy()
plt.plot(t_range.numpy(), predicted_curve, 'c-', label="Model Prediction")

# Training predictions
train_preds = seq_model(t_u.unsqueeze(1) / 100.0).detach().numpy()
plt.plot(t_u.numpy(), train_preds, 'kx', label="Training Predictions")

plt.legend()
plt.grid(True)
plt.show()
```
